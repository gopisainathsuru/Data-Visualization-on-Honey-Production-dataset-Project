### <font color='blue'>1. Import required libraries and read the dataset.</font>

```python
# importing required libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# reading dataset

data_df = pd.read_csv('honeyproduction.csv')
```

### <font color='blue'>3. Display the percentage distribution of the data in each year using the pie chart.</font>


```python
data_df.year.value_counts()             # to get distribution of data in every year as count
```

```python
# plotted pie chart using value_counts of year column

plt.pie(data_df.year.value_counts(), autopct = '%.1f%%',labels = data_df.year.value_counts().index)
plt.title('Percentage distribution of the data in each year')
plt.show()
```

### <font color='blue'>4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.</font>


```python
# plotted displot of priceperlb column data

sns.displot(data=data_df,x='priceperlb',kde=True)
plt.title('Distribution of the \"price per lb\"')
plt.show() 
```

## Inferences:
* the mode of the data likely to be near to 100
* priceperlb column data is positively skewed 


### <font color='blue'>5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.</font>


```python
# plotted scatter plot of numcol vs prodvalue columns data

sns.scatterplot(data=data_df,x='numcol',y='prodvalue')
plt.title('\'numcol\' and \'prodvalue\' relationship')
plt.show()
```

## Inferences:
* Number of Colonies and Production Value is linearly distributed
* numcol and prodvalue is positively correlated.

### <font color='blue'>6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.</font>


```python
plt.figure(figsize=(10,7.5))
sns.boxplot(data=data_df,x='year',y='prodvalue')
plt.title('Boxplot using \'year\',\'prodvalue\'')
plt.show()
```

## Inferences:
* The Production Value is more in the year 2010
* In all years, we can see that majority of prodvalue data present in the same range approximately.

### <font color='blue'>7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')</font>


```python
data_df1 = data_df[['numcol','yieldpercol','totalprod','prodvalue','year']]
sns.pairplot(data=data_df1,hue='year')
plt.show
```

## Inferences:
* numcol is positively correlated with totalprod and prodvalue
* totalprod is positively correlated with prodvalue and numcol
* prodvalue is positively correlated with numcol and totalprod


### <font color='blue'>8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol','total prod', 'stocks', 'price per lb', 'prodvalue')</font>

```python
data_df2=data_df[['numcol','yieldpercol','totalprod','stocks','priceperlb','prodvalue']]
data_df2.corr()
```

```python
sns.heatmap(data_df2.corr(),annot=True,linewidths=1)
plt.title('Correlation Matrix')
plt.show()
```

## Inferences:
